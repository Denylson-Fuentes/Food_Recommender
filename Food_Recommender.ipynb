```python
import numpy as np
import pandas as pd
import seaborn as sns
import sys
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Food Recommender/RAW_recipes.csv')
```

```python
df.head()
```

```python
## create new columns to expand the nutrution column

df[['calories', 'total fat (PDV)', 
    'sugar (PDV)', 'sodium (PDV)', 
    'protein (PDV)', 'saturated fat (PDV)', 'carbohydrates (PDV)']] = df.nutrition.str.split(',', expand = True)
```

```python
## replace any unnecessary values with and empty space

df['calories'] = df['calories'].apply(lambda x: x.replace('[', ''))
```

```python
df['carbohydrates (PDV)'] = df['carbohydrates (PDV)'].apply(lambda x: x.replace(']', ''))
```

```python
## Type cast column values for easier manipulation
df[['calories', 'total fat (PDV)', 
    'sugar (PDV)', 'sodium (PDV)', 
    'protein (PDV)', 'saturated fat (PDV)', 'carbohydrates (PDV)']] = df[['calories', 'total fat (PDV)', 
    'sugar (PDV)', 'sodium (PDV)', 
    'protein (PDV)', 'saturated fat (PDV)', 'carbohydrates (PDV)']].astype('float')
```

```python
df.dtypes
```

```python
df.head()
```

```python
#Drop all unnecessary rows that will just act as noise
df.drop(['id', 'contributor_id', 'submitted', 'tags', 'nutrition'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['food types'] = np.nan
df['food types'] = df['food types'].astype('str')
```

```python
## Identifying if the desserts are Vegan or not

for i in df['ingredients'].index:
    if('eggs' not in df['ingredients'][i]):
        if('ice-cream' in df['ingredients'][i] or 'chocolate' in df['ingredients'][i] or 'cookies' in df['ingredients'][i]):
            df['food types'][i] = 'Veg dessert'
            
    elif('eggs'  in df['ingredients'][i]):
        if('ice-cream' in df['ingredients'][i] or 'chocolate' in df['ingredients'][i] or 'cookies' in df['ingredients'][i]):
            df['food types'][i] = 'Non-Veg dessert'
```

```python
## Now we will be checking if any of the recipes have a calorie count less than 500 
## Those will be considered healthy

for i in df.index:
    if(df['food types'][i] != 'Veg dessert' and df['food types'][i] != 'Non-Veg dessert' and 20<df['calories'][i]<500):
        df['food types'][i] = 'Healthy'
```

```python
## Now we will check all other meals and determine if they are Vegan or not 
## This will depend on the meals having any meat

if i in df.index:
    if(df['food types'][i] != 'Veg dessert' and df['food types'][i] != 'Non-Veg dessert' and df['food types'][i] != 'Healthy'):
        if('chicken' in df['ingredients'][i] or 
           'eggs' in df['ingredients'][i] or 
           'ham' in df['ingredients'][i] or 
           'pepperoni' in df['ingredients'][i]):
            
            df['food types'][i] = 'Non-veg'
        else:
            df['food types'][i] = 'Veg'
```

```python
## Creating one hot encoding for the food types

types = pd.get_dummies(df['food types'])
```

```python
## Merging both dataset and new column to create one dataset with food types

df1 = pd.concat([df, types], axis=1)
```
